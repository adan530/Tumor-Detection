```python
from ultralytics import YOLO
from ultralytics import SAM

# Load the YOLO model
model = YOLO("/content/runs/detect/train/weights/best.pt") # pretrained YOLO model

# Run batched inference on a list images
results = model("/content/drive/MyDrive/test_images/meningioma_3.jpg") # return a list of Results objects

# Load SAM model
sam_model = SAM("sam2_b.pt")

for result in results:
  class_ids = result.boxes.cls.int().tolist() #noqa
  if len(class_ids):
      boxes = result.boxes.xyxy # Boxes objects for bbox outputs
      sam_results = sam_model(result.orig_img, bboxes=boxes, verbose=False, save=True, device=0)

# => Loop is used to get bounding boxes and get and assign class_ids, and implement SAM2
# => Arguments of sam_model bboxes(puts mask on the boxes if not assigned it will mask whole image)
```

Instance Segmentation using Yolo11 and Sam12
Instance Segmentation Model is putting mask on the custom object
Example: putting mask on sofa, table chair

Step1: Train model on the custom data object using dataset
=> Images and annotations (already mapped cordinates that defines the image)
==> Instance segmentation vs Object Detection Model
IS is more time consuming as we have to annotate precissely coordinate the image data while in object detection we just highlight the image only (like screenshot)
=> Solution to train your instance segmentation model without making custom data set
Yolo11 (for object detection)
Example: identify or make bounding box on  sofa,chair
Sam2 (Segment Anything to Model)  for instance segmentation
=> Meta model
=> output of Yolo11 will be input of Sam2 that is bounding boxes around the object
=> Sam2 will put mask on the bounding boxes around the object

Benefit:
Object Detection data preparation will be simple and quick for instance segmentation model
==> Custom Dataset => Object Detection Model Yolo11 => Sam2 will put mask on it
